```python
def get_velikost(v):
    split_v = str(v).split()
    try:
        return float(split_v[0].replace(',', '.'))
    except ValueError:
        return None

def get_cena(c):
    c = str(c).split('€')[0].split('EUR')[0].replace('.', '').replace(',', '.').strip()

    try:
        return float(c)
    except ValueError:
        return None

def očisti_ceno(cena_str, velikost):
    if pd.isna(cena_str):
        return np.nan
    cena_str = str(cena_str).lower().replace("€", "").replace("€", "").replace(".", "").strip()
    
    try:
        if "m2" in cena_str:
            vrednost_na_m2 = float(cena_str.replace("m2", "").strip())
            return vrednost_na_m2 * velikost
        else:
            return float(cena_str)
    except:
        return np.nan
```

```python
hs_df = kvadrati_gradnje_df[(kvadrati_gradnje_df['vrsta'] == "Stanovanje") | (kvadrati_gradnje_df['vrsta'] == "Hiša")]
hs_df = hs_df.copy()
hs_df['velikost_clean'] = hs_df['velikost'].apply(get_velikost)
# hs_df['cena_clean'] = hs_df['cena'].apply(get_cena)
hs_df['cena_clean'] = hs_df.apply(lambda row: očisti_ceno(row['cena'], row['velikost']), axis=1)
hs_df['cena_na_m2'] = hs_df['cena_clean'] / hs_df['velikost_clean']
hs_df
```

```python
cena_na_m2 = hs_df["cena_na_m2"].dropna()

dist = distfit()
dist.fit_transform(cena_na_m2)
```

Zgleda je Beta porazdelitev.

```python
# leto_normalized = (leto_gradnje - leto_gradnje.min() + random.randint(1,100)/100) / (leto_gradnje.max() - leto_gradnje.min())
# print(leto_normalized)
pars_b = beta.fit(cena_na_m2)
print(pars_b)

xb = np.linspace(cena_na_m2.min(), cena_na_m2.max(), 1000)
B_fit = [beta.pdf(x, *pars_b) for x in xb]

plt.figure()
plt.hist(cena_na_m2, bins="sqrt", density=True)
# plt.hist(leto_normalized)
plt.plot(xb, B_fit, label="Beta porazdelitev", linewidth=2.0)
plt.title('Histogram cene nepremičnin')
plt.xlabel('Cena na m2')
plt.ylabel('Število nepremičnin (normalizirano)')
plt.legend();
```

```python
import scipy.stats as stats

plt.figure(figsize=(8, 5))
stats.probplot(cena_na_m2, dist="beta", sparams=pars_b, plot=plt)
plt.title("Beta porazdelitev")

plt.show()
```

```python
# Funkcija za pridobitev vrednosti iz 'splosno'
def extract_energy(row):
    try:
        entries = str(row).split('|')
        for entry in entries:
            if '=' in entry:
                key, value = entry.split('=', 1)
                if key.strip().lower().startswith('ener'):
                    return value.strip()
    except:
        return None

hs_df5 = hs_df.copy()
hs_df5['energetski_izkaz'] = hs_df5['splosno'].apply(extract_energy)
hs_df5
```

```python
# Definiran vrstni red razredov (A1 in A2 boljši kot B, C, ...)
energy_order = ['A1', 'A2', 'B1', 'B2', 'C', 'D', 'E', 'F', 'G']
# energy_order = ['G', 'F', 'E', 'D', 'C', 'B2', 'B1', 'A2', 'A1']
energy_mapping = {'G':255 , 'F':180, 'E':128, 'D':82.5, 'C':47.5, 'B2':30, 'B1':20, 'A2':12.5, 'A1':5}
# energy_mapping = {val: idx for idx, val in energy_order}

# Pretvori v številke
hs_df5['energetski_izkaz_ordinal'] = hs_df5['energetski_izkaz'].map(energy_mapping)
hs_df5
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtriramo veljavne vrstice
valid_energy = hs_df5.dropna(subset=['leto_gradnje', 'energetski_izkaz_ordinal'])

plt.figure(figsize=(10, 6))
# sns.scatterplot(data=valid_energy, x='leto_gradnje', y='energetski_izkaz_ordinal',
#                 alpha=0.6, color='purple')
# sns.regplot(data=valid_energy, x='leto_gradnje', y='energetski_izkaz_ordinal', scatter_kws={"alpha": 0.5}, line_kws={"color": "red", "alpha": 0.8}, ci=None)
sns.regplot(data=valid_energy, x='leto_gradnje', y='energetski_izkaz_ordinal',
            scatter_kws={"alpha": 0.5}, line_kws={"color": "red", "alpha": 0.8}, ci=None)

print(energy_mapping.values())
# Prilagodi y-tick labele, da pokaže razrede, ne številk
ticks = list(reversed(energy_mapping.values()))
labels = energy_order
plt.yticks(ticks=ticks, labels=labels)

plt.title('Energetski razredi glede na leto gradnje', fontsize=14)
plt.xlabel('Leto gradnje', fontsize=12)
plt.ylabel('Energetski razred', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlim(1800)
plt.ylim(-1, 290)
plt.tight_layout()
plt.show()
```
